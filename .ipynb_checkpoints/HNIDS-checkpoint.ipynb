```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import random
import multiprocessing
from joblib import Parallel, delayed
```

```python

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

class HNIDS:
    """
    Hybrid Network-based Intrusion Detection System using EGA-PSO (Enhanced Genetic Algorithm with 
    Particle Swarm Optimization) and IRF (Improved Random Forest).
    """
    
    def __init__(self, population_size=50, max_iterations=30, crossover_rate=0.8, 
                 mutation_rate=0.1, w=0.7, c1=1.5, c2=1.5, seed=42):
        """
        Initialize the HNIDS model with hyperparameters.
        
        Args:
            population_size: Size of the initial population for GA
            max_iterations: Maximum number of iterations for EGA-PSO
            crossover_rate: Probability of crossover for GA
            mutation_rate: Probability of mutation for GA
            w: Inertia weight for PSO
            c1: Cognitive coefficient for PSO
            c2: Social coefficient for PSO
            seed: Random seed for reproducibility
        """
        self.population_size = population_size
        self.max_iterations = max_iterations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        self.w = w
        self.c1 = c1
        self.c2 = c2
        self.seed = seed
        self.best_features = None
        self.irf_model = None
        self.feature_names = None
        self.categorical_encoders = {}
        np.random.seed(seed)
        random.seed(seed)
        
    def _preprocess_data(self, data, is_training=False):
        """
        Preprocess the data with Min-Max normalization.
        
        Args:
            data: Input DataFrame with features and target ('label' column)
            is_training: Whether this is training data (to fit transformers) or testing data
        
        Returns:
            Normalized DataFrame of features only (target is kept separately)
        """
        # Make a copy to avoid modifying the original data
        data_copy = data.copy()
        
        # Extract target if it exists
        y = None
        if 'label' in data_copy.columns:
            y = data_copy['label'].copy()
            data_copy = data_copy.drop('label', axis=1)
        
        # Handle categorical features
        categorical_columns = [col for col in data_copy.columns if data_copy[col].dtype == 'object']
        
        # For categorical columns, use one-hot encoding
        if categorical_columns:
            if is_training:
                # For training data, fit and transform
                for col in categorical_columns:
                    # Get all unique values in this column
                    unique_values = data_copy[col].unique()
                    # Create a dictionary to map values to one-hot columns
                    self.categorical_encoders[col] = unique_values
                    
                    # Create one-hot encoded columns
                    for val in unique_values:
                        new_col = f"{col}_{val}"
                        data_copy[new_col] = (data_copy[col] == val).astype(int)
                        
                    # Drop the original column
                    data_copy = data_copy.drop(col, axis=1)
            else:
                # For test data, use the encoders from training
                for col in categorical_columns:
                    if col in self.categorical_encoders:
                        # Create columns for all values seen during training
                        for val in self.categorical_encoders[col]:
                            new_col = f"{col}_{val}"
                            if col in data_copy.columns:  # Ensure column exists
                                data_copy[new_col] = (data_copy[col] == val).astype(int)
                            else:
                                data_copy[new_col] = 0  # If column doesn't exist, set all to 0
                        
                        # Drop the original column if it exists
                        if col in data_copy.columns:
                            data_copy = data_copy.drop(col, axis=1)
                            
        # Apply Min-Max normalization
        numeric_columns = data_copy.select_dtypes(include=['number']).columns
        if len(numeric_columns) > 0:
            scaler = MinMaxScaler()
            data_copy[numeric_columns] = scaler.fit_transform(data_copy[numeric_columns])
        
        # Store feature names if this is training data
        if is_training:
            self.feature_names = data_copy.columns.tolist()
        # For test data, ensure all expected columns are present
        elif self.feature_names:
            # Add missing columns with zeros
            for col in self.feature_names:
                if col not in data_copy.columns:
                    data_copy[col] = 0
            # Reorder columns to match training order
            data_copy = data_copy[self.feature_names]
        
        return data_copy
    
    def _initialize_population(self, n_features):
        """
        Initialize the binary population for GA.
        
        Args:
            n_features: Number of features in the dataset
        
        Returns:
            Initial population represented as binary vectors
        """
        return np.random.randint(0, 2, size=(self.population_size, n_features))
    
    def _calculate_fitness(self, chromosome, X, y):
        """
        Calculate fitness value for a chromosome using IRF.
        
        Args:
            chromosome: Binary vector representing feature selection
            X: Feature matrix
            y: Target vector
        
        Returns:
            Fitness value based on IRF performance
        """
        # Select features based on chromosome
        if sum(chromosome) == 0:  # If no features selected
            return 0.0
        
        selected_features = X.iloc[:, np.where(chromosome == 1)[0]]
        
        # Use IRF for fitness evaluation
        try:
            X_train, X_val, y_train, y_val = train_test_split(
                selected_features, y, test_size=0.2, random_state=self.seed, stratify=y
            )
            
            # Create an Improved Random Forest model
            irf = RandomForestClassifier(
                n_estimators=100,
                max_depth=10,
                min_samples_split=4,
                min_samples_leaf=2,
                criterion='gini',
                random_state=self.seed,
                n_jobs=-1,
                class_weight='balanced'
            )
            
            irf.fit(X_train, y_train)
            y_pred = irf.predict(X_val)
            
            # Calculate fitness as a combination of performance metrics
            accuracy = accuracy_score(y_val, y_pred)
            precision = precision_score(y_val, y_pred, average='weighted')
            recall = recall_score(y_val, y_pred, average='weighted')
            f1 = f1_score(y_val, y_pred, average='weighted')
            
            # Penalize for too many features to achieve feature reduction
            feature_penalty = 0.2 * (sum(chromosome) / len(chromosome))
            
            # Combined fitness function as per paper - Section 4.3.2
            fitness_value = (0.4 * accuracy + 0.2 * precision + 0.2 * recall + 0.2 * f1) - feature_penalty
            
            return fitness_value
            
        except Exception as e:
            print(f"Error in fitness calculation: {e}")
            return 0.0
    
    def _enhanced_selection(self, population, fitness_values):
        """
        Enhanced selection process for GA as described in the paper.
        
        Args:
            population: Current population
            fitness_values: Fitness values for each individual
        
        Returns:
            Selected parents and rejected individuals
        """
        # Calculate selection probability as per equation (7)
        selection_probs = fitness_values / np.sum(fitness_values)
        
        # Apply enhanced selection with size of 0.264 as mentioned in the paper
        selected_size = max(2, int(0.264 * self.population_size))
        
        # Select individuals based on probability
        selected_indices = np.random.choice(
            range(self.population_size), 
            size=selected_size, 
            replace=False, 
            p=selection_probs
        )
        
        # Find rejected individuals (to be processed by PSO)
        rejected_indices = np.setdiff1d(range(self.population_size), selected_indices)
        
        selected_parents = population[selected_indices]
        rejected_individuals = population[rejected_indices]
        
        return selected_parents, rejected_individuals
    
    def _enhanced_crossover(self, parents):
        """
        Enhanced crossover process for GA.
        
        Args:
            parents: Selected parents
        
        Returns:
            Offspring after crossover
        """
        n_parents = len(parents)
        n_features = parents.shape[1]
        offspring = []
        
        # Ensure even number of parents for pairing
        if n_parents % 2 != 0:
            parents = np.vstack([parents, parents[0]])
            n_parents += 1
            
        # Shuffle parents for random pairing
        np.random.shuffle(parents)
        
        for i in range(0, n_parents, 2):
            parent1 = parents[i]
            parent2 = parents[i+1]
            
            if np.random.rand() < self.crossover_rate:
                # Select a random crossover point
                crossover_point = np.random.randint(1, n_features)
                
                # Perform crossover
                child1 = np.concatenate([parent1[:crossover_point], parent2[crossover_point:]])
                child2 = np.concatenate([parent2[:crossover_point], parent1[crossover_point:]])
                
                offspring.append(child1)
                offspring.append(child2)
            else:
                # Keep parents unchanged
                offspring.append(parent1)
                offspring.append(parent2)
                
        return np.array(offspring)
    
    def _enhanced_mutation(self, offspring):
        """
        Enhanced mutation with Linear Discriminant Method (LDM) as described in the paper.
        
        Args:
            offspring: Offspring after crossover
        
        Returns:
            Mutated offspring
        """
        n_features = offspring.shape[1]
        
        for i in range(len(offspring)):
            for j in range(n_features):
                if np.random.rand() < self.mutation_rate:
                    # Flip the bit
                    offspring[i, j] = 1 - offspring[i, j]
                    
        return offspring
    
    def _pso_update(self, rejected_individuals, X, y, global_best_position):
        """
        PSO method to update rejected individuals (non-fit participants).
        
        Args:
            rejected_individuals: Individuals rejected in GA selection
            X: Feature matrix
            y: Target vector
            global_best_position: Global best position found so far
        
        Returns:
            Updated individuals after PSO
        """
        if len(rejected_individuals) == 0:
            return np.array([])
            
        n_rejected = len(rejected_individuals)
        n_features = rejected_individuals.shape[1]
        
        # Initialize velocity
        velocity = np.zeros((n_rejected, n_features))
        
        # Initialize personal best
        personal_best_position = rejected_individuals.copy()
        personal_best_fitness = np.array([
            self._calculate_fitness(rejected_individuals[i], X, y) 
            for i in range(n_rejected)
        ])
        
        # Update rejected individuals using PSO
        new_positions = rejected_individuals.copy().astype(float)
        
        for i in range(n_rejected):
            r1, r2 = np.random.rand(2)
            
            # Update velocity
            velocity[i] = (self.w * velocity[i] + 
                          self.c1 * r1 * (personal_best_position[i] - rejected_individuals[i]) + 
                          self.c2 * r2 * (global_best_position - rejected_individuals[i]))
            
            # Update position
            new_positions[i] = new_positions[i] + velocity[i]
            
            # Convert to binary using sigmoid function
            sigmoid = 1 / (1 + np.exp(-new_positions[i]))
            new_binary_position = (sigmoid > 0.5).astype(int)
            
            # Calculate fitness for the new position
            new_fitness = self._calculate_fitness(new_binary_position, X, y)
            
            # Update personal best if needed
            if new_fitness > personal_best_fitness[i]:
                personal_best_position[i] = new_binary_position
                personal_best_fitness[i] = new_fitness
                
        # Convert final positions to binary
        updated_positions = (1 / (1 + np.exp(-new_positions)) > 0.5).astype(int)
        
        return updated_positions
    
    def fit(self, X, y):
        """
        Fit the HNIDS model to the data.
        
        Args:
            X: Feature matrix
            y: Target vector
        
        Returns:
            Self (fitted model)
        """
        # Create a dataframe with both features and target
        data = pd.concat([X, pd.DataFrame(y, columns=['label'])], axis=1)
        
        # Preprocess data - flag as training data to fit encoders
        X_normalized = self._preprocess_data(data, is_training=True)
        
        n_features = X_normalized.shape[1]
        
        # Initialize population
        population = self._initialize_population(n_features)
        
        # Calculate initial fitness values
        fitness_values = np.array([
            self._calculate_fitness(population[i], X_normalized, y) 
            for i in range(self.population_size)
        ])
        
        # Track the best solution
        best_idx = np.argmax(fitness_values)
        global_best_position = population[best_idx].copy()
        global_best_fitness = fitness_values[best_idx]
        
        # Main EGA-PSO loop
        for iteration in range(self.max_iterations):
            # Enhanced selection
            selected_parents, rejected_individuals = self._enhanced_selection(population, fitness_values)
            
            # Enhanced crossover
            offspring = self._enhanced_crossover(selected_parents)
            
            # Enhanced mutation
            offspring = self._enhanced_mutation(offspring)
            
            # PSO update for rejected individuals
            updated_rejected = self._pso_update(rejected_individuals, X_normalized, y, global_best_position)
            
            # Combine updated individuals
            if len(updated_rejected) > 0:
                new_population = np.vstack([offspring, updated_rejected])
            else:
                new_population = offspring
                
            # Ensure population size is maintained
            if len(new_population) > self.population_size:
                new_population = new_population[:self.population_size]
            elif len(new_population) < self.population_size:
                # Add random individuals if needed
                additional = self._initialize_population(n_features)
                new_population = np.vstack([new_population, additional[:self.population_size-len(new_population)]])
            
            # Update population and fitness values
            population = new_population
            fitness_values = np.array([
                self._calculate_fitness(population[i], X_normalized, y) 
                for i in range(self.population_size)
            ])
            
            # Update global best
            current_best_idx = np.argmax(fitness_values)
            if fitness_values[current_best_idx] > global_best_fitness:
                global_best_position = population[current_best_idx].copy()
                global_best_fitness = fitness_values[current_best_idx]
                
            print(f"Iteration {iteration+1}/{self.max_iterations}, Best Fitness: {global_best_fitness:.4f}, "
                  f"Selected Features: {sum(global_best_position)}/{n_features}")
        
        # Save the best features
        self.best_features = global_best_position
        
        # Train the final IRF model on selected features
        selected_columns = X_normalized.columns[np.where(self.best_features == 1)[0]]
        X_selected = X_normalized[selected_columns]
        
        # Create and fit the IRF model
        self.irf_model = RandomForestClassifier(
            n_estimators=1000,
            max_depth=10,
            min_samples_split=4,
            min_samples_leaf=2,
            criterion='gini',
            random_state=self.seed,
            n_jobs=-1,
            class_weight='balanced'
        )
        
        self.irf_model.fit(X_selected, y)
        
        return self
    
    def predict(self, X):
        """
        Make predictions on new data.
        
        Args:
            X: Feature matrix for prediction
        
        Returns:
            Predicted labels
        """
        if self.best_features is None or self.irf_model is None:
            raise ValueError("Model has not been fitted yet. Call fit() first.")
        
        # Preprocess data - pass is_training=False to use encoders from training
        X_normalized = self._preprocess_data(X, is_training=False)
        
        # Select features using best_features mask
        if len(self.best_features) != X_normalized.shape[1]:
            # If feature dimensions don't match, use all features as fallback
            print(f"Warning: Feature dimensions mismatch. Expected {len(self.best_features)}, got {X_normalized.shape[1]}.")
            X_selected = X_normalized
        else:
            # Use selected features based on best_features mask
            selected_indices = np.where(self.best_features == 1)[0]
            selected_columns = X_normalized.columns[selected_indices]
            X_selected = X_normalized[selected_columns]
        
        # Make predictions
        return self.irf_model.predict(X_selected)
    
    def evaluate(self, X, y, binary_classification=True):
        """
        Evaluate the model on test data.
        
        Args:
            X: Feature matrix
            y: True labels
            binary_classification: Boolean indicating if binary or multi-class classification
        
        Returns:
            Dictionary with evaluation metrics
        """
        y_pred = self.predict(X)
        
        # Calculate metrics
        metrics = {
            'accuracy': accuracy_score(y, y_pred),
            'precision': precision_score(y, y_pred, average='weighted'),
            'recall': recall_score(y, y_pred, average='weighted'),
            'f1_score': f1_score(y, y_pred, average='weighted'),
            'confusion_matrix': confusion_matrix(y, y_pred)
        }
        
        print("Classification Report:")
        print(classification_report(y, y_pred))
        
        print("Confusion Matrix:")
        cm = confusion_matrix(y, y_pred)
        
        # Visualize confusion matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title('Confusion Matrix')
        plt.show()
        
        # Calculate false alarm rate and other metrics
        if binary_classification:
            tn, fp, fn, tp = cm.ravel()
            metrics['tpr'] = tp / (tp + fn) * 100  # True Positive Rate
            metrics['tnr'] = tn / (tn + fp) * 100  # True Negative Rate
            metrics['fpr'] = fp / (fp + tn) * 100  # False Positive Rate
            metrics['fnr'] = fn / (fn + tp) * 100  # False Negative Rate
            metrics['false_alarm'] = fp / (fp + tn) * 100  # False Alarm
            
            print(f"TPR: {metrics['tpr']:.2f}%")
            print(f"TNR: {metrics['tnr']:.2f}%")
            print(f"FPR: {metrics['fpr']:.2f}%")
            print(f"FNR: {metrics['fnr']:.2f}%")
            print(f"False Alarm Rate: {metrics['false_alarm']:.2f}%")
            
        return metrics

# Example usage:
def load_and_preprocess_nsl_kdd():
    """
    Load and preprocess the NSL-KDD dataset.
    
    Returns:
        X_train, X_test, y_train, y_test: Split dataset
    """
    # NOTE: You need to download the NSL-KDD dataset from the following link:
    # https://www.unb.ca/cic/datasets/nsl.html
    
    try:
        # Load training data - read with flexible handling of columns
        train_data = pd.read_csv('KDDTrain+.txt', header=None)
        test_data = pd.read_csv('KDDTest+.txt', header=None)
        
        # Determine the actual number of columns
        n_columns = train_data.shape[1]
        print(f"Detected {n_columns} columns in the NSL-KDD dataset")
        
        # Features for NSL-KDD dataset - using standard feature names
        # The last column is 'label' (attack type) and second-to-last might be 'difficulty_level'
        base_columns = [
            'duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 
            'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in', 
            'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations', 
            'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login', 
            'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 
            'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate', 
            'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 
            'dst_host_same_srv_rate', 'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate', 
            'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate', 
            'dst_host_rerror_rate', 'dst_host_srv_rerror_rate'
        ]
        
        # Adjust columns based on detected number
        if n_columns == 42:
            columns = base_columns + ['label']
        elif n_columns == 43:
            columns = base_columns + ['difficulty_level', 'label']
        else:
            # Generate generic column names if structure is different
            columns = [f'feature_{i}' for i in range(n_columns-1)] + ['label']
            
        train_data.columns = columns
        test_data.columns = columns
        
        # Convert labels to binary for binary classification (taking care of different label formats)
        def to_binary_label(label):
            if isinstance(label, str):
                return 0 if label.lower() == 'normal' else 1
            else:
                return 0 if label == 0 else 1  # For numeric labels, assuming 0 is normal
                
        train_data['binary_label'] = train_data['label'].apply(to_binary_label)
        test_data['binary_label'] = test_data['label'].apply(to_binary_label)
        
        # Drop difficulty_level if it exists
        if 'difficulty_level' in train_data.columns:
            X_train = train_data.drop(['label', 'binary_label', 'difficulty_level'], axis=1)
            X_test = test_data.drop(['label', 'binary_label', 'difficulty_level'], axis=1)
        else:
            X_train = train_data.drop(['label', 'binary_label'], axis=1)
            X_test = test_data.drop(['label', 'binary_label'], axis=1)
            
        y_train = train_data['binary_label']
        y_test = test_data['binary_label']
        
        return X_train, X_test, y_train, y_test
        
    except FileNotFoundError:
        print("NSL-KDD dataset files not found. Please download from https://www.unb.ca/cic/datasets/nsl.html")
        print("Using sample data instead for demonstration.")
        
        # Generate dummy data for demonstration
        from sklearn.datasets import make_classification
        
        X, y = make_classification(
            n_samples=1000, 
            n_features=41, 
            n_informative=10, 
            n_redundant=10, 
            n_classes=2, 
            random_state=42
        )
        
        X_train, X_test, y_train, y_test = train_test_split(
            pd.DataFrame(X), pd.Series(y), test_size=0.2, random_state=42
        )
        
        return X_train, X_test, y_train, y_test

def main():
    """Main function to demonstrate HNIDS."""
    print("Loading and preprocessing NSL-KDD dataset...")
    X_train, X_test, y_train, y_test = load_and_preprocess_nsl_kdd()
    
    print(f"Training data shape: {X_train.shape}")
    print(f"Testing data shape: {X_test.shape}")
    
    # Create and fit HNIDS model
    print("\nInitializing HNIDS model...")
    hnids = HNIDS(
        population_size=30,  # Reduced for demonstration
        max_iterations=10,   # Reduced for demonstration
        crossover_rate=0.8,
        mutation_rate=0.1,
        w=0.7,
        c1=1.5,
        c2=1.5,
        seed=42
    )
    
    print("\nFitting HNIDS model...")
    hnids.fit(X_train, y_train)
    
    print("\nEvaluating HNIDS model on test data...")
    metrics = hnids.evaluate(X_test, y_test, binary_classification=True)
    
    print("\nSummary of Results:")
    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"Precision: {metrics['precision']:.4f}")
    print(f"Recall: {metrics['recall']:.4f}")
    print(f"F1-score: {metrics['f1_score']:.4f}")
    
    # Show selected features
    if hasattr(hnids, 'best_features') and hnids.best_features is not None:
        # Use feature names from the model instead of X_train columns
        # which may differ after preprocessing
        if hasattr(hnids, 'feature_names') and hnids.feature_names is not None:
            feature_indices = np.where(hnids.best_features == 1)[0]
            valid_indices = [idx for idx in feature_indices if idx < len(hnids.feature_names)]
            
            if valid_indices:
                selected_features = [hnids.feature_names[idx] for idx in valid_indices]
                print(f"\nSelected Features ({len(selected_features)}/{len(hnids.feature_names)}):")
                print(selected_features)
            else:
                print("\nNo features were selected.")
        else:
            print("\nFeature names not available.")
    else:
        print("\nNo features were selected.")

if __name__ == "__main__":
    main()
```
